```python
from pyecharts import Geo
from pyecharts import Map
from pyecharts import Bar,Line,Overlap
import charts
```

```python
add =["北京","重庆","上海","天津","河南","广东","山东","安徽","四川","江苏","河北","湖北","陕西","湖南","山西","浙江","甘肃","广西","江西","福建","辽宁","贵州","云南","黑龙江","内蒙古","吉林","新疆","宁夏","海南","青海","西藏"]
num = [6.3,25,0,5.5,98.3,75.8,59.2,49.9,62,33.1,48.6,37.4,31.9,45.1,30.5,30.6,27.3,40,38,20.0,18.5,44,30,19.0,0,15.0,20.7,6.95,5.88,0,2.53]
# print(add)
# print(num)
dict_data = dict(zip(add,num))
data = [tuple(i) for i in zip(add,num)]
print(data)
```

```python
data_pro = data

geo = Map("各省高考人数分布", "data from : gaokao.eol.cn", title_color="#fff",  
          title_pos="center", width=1000,  
          height=600, background_color='#404a59')  
attr, value = geo.cast(data_pro)
geo.add("", attr, value, visual_range=[0, 80],maptype='china',visual_text_color="#fff",  
        symbol_size=10, is_visualmap=True)  
geo.render("各省高考人数分布.html")#生成html文件  
geo#直接在notebook中显示  
```

```python
# 高考人数
gaokao_num = [940,940,942,939,912,915,933,946,1020,1050,1010,950,877,729,613,510,454,375]
gaokao_num.reverse()
print(gaokao_num)
# 录取人数
luqu_num = [700,705,700,698,694,685,675,657,629,599,566,546,504,447,382,320,268,221]
luqu_num.reverse()
print(luqu_num)
# 录取率
luqu_lev= [74.46,75,74.3,74.3,76,75,72,69,62,57,56,57,57,61,62,63,59,59]
luqu_lev.reverse()
print(luqu_lev)
```

```python
# 年份
attr = ["{}".format(i) for i in range(2000, 2018)]
print(attr)
bar = Bar()
bar.add("高考人数(万人)", attr, gaokao_num)
bar.add("录取人数(万人)", attr, luqu_num)
line = Line()
line.add('录取率',attr,luqu_lev,line_color='#90ed7d',line_width=4)
ol = Overlap("近十年高考数据")
ol.add(bar)
ol.add(line)
ol.show_config()
ol.render()
```

```python
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': '2000-2017年高考数据'},
    'subtitle': {'text': 'Source: edu.sina.com.cn'},
    'xAxis'   : {'categories': ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016','2017']},
    'yAxis'   : {'title': {'text': '万人/年'}},
#     'plotOptions':{'column':{'dataLabels':{'enable': True,'shadow':True}}}
    
    }
series =  [{
    'type': 'column',
    'name': '高考人数',
    'data': gaokao_num
},{
    'type': 'column',
    'name': '录取人数',
    'data': luqu_num
# },{
#     'type': 'line',
#     'name': '录取率',
#     'data': luqu_lev,
#     'marker':{
#         'lineWidth':2,
#         'lineColor':'#90ed7d',
#         'fillColor':'orange'
#     }
}
]

charts.plot(series, options=options, show='inline')
```
